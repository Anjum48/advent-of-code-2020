```python
import pandas as pd
from itertools import combinations
from pathlib import Path

INPUT_PATH = Path("input")
```

```python
df = pd.read_csv(INPUT_PATH / "day01_input.txt", header=None)
report = df.iloc[:, 0].tolist()
report
```

```python
# Part 1
for a, b in combinations(report, 2):
    if a + b == 2020:
        print(f"Pair: {a} + {b} = {a + b}. Product: {a*b}")
```

```python
# Part 2
for a, b, c in combinations(report, 3):
    if a + b + c == 2020:
        print(f"Triplet: {a} + {b} + {c}= {a + b + c}. Product: {a*b*c}")
```
