```python
import pandas as pd
import numpy as np
from pathlib import Path

INPUT_PATH = Path("input")
```

```python
df = pd.read_csv(INPUT_PATH / "day02_input.txt", header=None, delimiter=" |:|-", names=["Min_letters", "Max_letters", "Letter", "Gap", "Password"], engine="python")
del df["Gap"]
df
```

```python
def count_letters(x):
    word = x["Password"]
    letter = x["Letter"]
    return word.count(letter)

count_letters(df.iloc[0])
```

```python
df["Length"] = df["Password"].str.len()
df["Letter_counts"] = df.apply(count_letters, axis=1)
df
```

```python
df["Pass"] = np.where((df["Letter_counts"] >= df["Min_letters"]) & (df["Letter_counts"] <= df["Max_letters"]), 1, 0)
```

```python
df
```

```python
# Answer
df["Pass"].sum()
```

```python
def check_char1(x):
    pos = x["Min_letters"] - 1
    letter = x["Letter"]
    password = x["Password"]
    
    if password[pos] == letter:
        return 1
    else:
        return 0

    
def check_char2(x):
    pos = x["Max_letters"] - 1
    letter = x["Letter"]
    password = x["Password"]
    
    if password[pos] == letter:
        return 1
    else:
        return 0
```

```python
df["pass1"] = df.apply(check_char1, axis=1)
df["pass2"] = df.apply(check_char2, axis=1)
df["pass3"] = df["pass1"] + df["pass2"]
df["pass3"].value_counts()
```
